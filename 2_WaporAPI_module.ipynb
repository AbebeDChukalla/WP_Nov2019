### 2.2 Specify the shapefile of ROI
To request a cropped raster, raster_id and time_code of the raster is needed

```python
import shapefile

dir_mate =  "Materials"
shpfile  =  os.path.join(dir_mate, "Shapefile", "Xinavane_1.shp") 
shape    =  shapefile.Reader(shpfile)

bbox     =  shape.bbox  #[xmin,ymin,xmax,ymax], longitude and latitude
bbox
```

## 2b. Bulk-downloading
### 2b.I Download AETI

#### Specify the cube code of the product

```python
cube_code   = 'L2_AETI_D'
cube_code
```

#### Get information of a cube

```python
cube_info   = WaPOR.API.getCubeInfo(cube_code)
cube_info 
```

#### Multiplier

```python
multiplier1 = cube_info['measure']['multiplier']

print('Description: {0} \n\nmultiplier: {1} \n\nunit: {2} \n'.format(cube_info['description'],
                    cube_info['measure']['multiplier'],
                    cube_info['measure']['unit']))
multiplier1
```

#### Define the period for the download - between the available time range

```python
df_avail1 = WaPOR.API.getAvailData(cube_code,
                                time_range = '2009-01-01,2019-12-31')  #2019-09-30
df_avail1
```

#### Define the directory of the folder to save the raster data

```python
## Define the directory of the folder to save the AETI data

dir_proj = os.getcwd()  
dir_data = "Data"
# Dir = 'E:\WaPORwpDocker\Data'

foldername = '2a%s'%(cube_code)

Dir = os.path.join(dir_proj, dir_data, foldername)
Dir
```

#### Download the raster

```python
# make directory if it does not exit
if not os.path.exists(Dir):
    os.makedirs(Dir)

for index,row in df_avail1.iterrows():
    ### get download url
    download_url  = WaPOR.API.getCropRasterURL(bbox,cube_code,
                                           row['time_code'],
                                           row['raster_id'],
                                           WaPOR.API.Token,
                                           print_job=False)
    outfilename   = os.path.join(Dir, '{0}.tif'.format(row['raster_id']))
    download_file = os.path.join(Dir,'{0}.tif'.format(row['raster_id']))
    ### Download raster file
    resp = requests.get(download_url) 
    open(download_file,'wb').write(resp.content) 

    ### number of days
    timestr = row['time_code']
    startdate = datetime.datetime.strptime(timestr[1:11],'%Y-%m-%d')
    enddate = datetime.datetime.strptime(timestr[12:22],'%Y-%m-%d')
    ndays = (enddate.timestamp()-startdate.timestamp())/86400

    ### correct raster with multiplier and number of days in dekad
    driver, NDV, xsize, ysize, GeoT, Projection= gis.GetGeoInfo(download_file)
    Array = gis.OpenAsArray(download_file,nan_values=True)
    Array = np.where(Array<0,0,Array) #mask out flagged value -9998
    CorrectedArray = Array*multiplier1*ndays                   
    gis.CreateGeoTiff(outfilename,CorrectedArray,
                      driver, NDV, xsize, ysize, GeoT, Projection)
```
